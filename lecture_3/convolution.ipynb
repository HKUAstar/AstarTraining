```python
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
```

```python
def show(image):
    plt.imshow(image)
    plt.show()
```

```python
def convolve(array, kernel):
    result = []
    m, n = array.shape
    p, q = kernel.shape
    for i in range(m - p + 1):
        current_row = []
        for j in range(n - q + 1):
            current_sum = 0
            for k in range(p):
                for l in range(q):
                    current_sum += array[i + k][j + l] * kernel[k][l]
            current_row.append(current_sum)
        result.append(current_row)
    return np.array(result)
```

```python
image = Image.open('./robot.png')
image = np.asarray(image)
image = image[::3, ::3, -1]
show(image)
```

```python
kernel = 1 / 6 * np.array([
    [0, 0, 0],
    [0, 1, 0],
    [0, 0, 0]
])
new_image = convolve(image, kernel)
show(new_image)
```
